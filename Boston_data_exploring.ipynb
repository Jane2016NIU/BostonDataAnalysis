# 2 Price correlation

2.1 For each of the features you analyzed above, generate a scatter plot of the feature vs.
price and report the Pearson coefficient. The plot should be clearly labelled and titled. 

```python
import numpy as np
from sklearn import datasets
from scipy.stats import pearsonr
from scipy.stats import scoreatpercentile
from matplotlib import pyplot as plt

boston = datasets.load_boston()

def plot_data(x, y, name):
    plt.plot(x, y, ".")
    plt.ylabel("Price in thousands")
    plt.xlabel(name)
    plt.title(name + " vs. Price")
    plt.show()

y = np.array(boston['target'])
ar = np.array(boston['data'])  
for i, name in enumerate(boston['feature_names']):
    print(name)
    x = ar[:,i]
    plot_data(x, y, name)
    print('Pearson’s correlation coefficient,2-tailed p-value')
    print(pearsonr(x,y))
    print('\n\n')


```
